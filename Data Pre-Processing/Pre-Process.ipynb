```python
# unzip the the dataset
import zipfile as zf
files = zf.ZipFile("bbcsport-fulltext.zip", 'r')
files.extractall('BBCSports')
files.close()
```

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# give the proper path for the documents directories
data_folder = 'C:/Users/Dell/Downloads/New folder/bbcsport'

folders = ["tennis","rugby","football","cricket","athletics"]
```

```python
os.chdir(data_folder)
```

```python
# x for text and y for labels
x = []
y = []
```

```python
# read all the documents at the given path and remove newlines and blank lines and append the class label
for i in folders:
    files = os.listdir(i)
    for text_file in files:
        file_path = i + "/" +text_file
        print ("reading file:", file_path)
        with open(file_path) as f:
            currentline = ""
            for line in f:
                if line.startswith('>'):
                    line = line.rstrip('\n')
                    if currentline != "": x.append (currentline)
                    x.append (line)
                    currentline = ""
                else:
                    line = line.rstrip('\n')
                    currentline = currentline + line
            x.append (currentline)
            y.append(i)
            f.close()
 
```

```python
data = {'text': x, 'label': y}       
df = pd.DataFrame(data)
print ('writing csv flie ...')
df.to_csv('C:/Users/Dell/Downloads/New folder/bbcsport/BCC_Sports.csv', index=False)
```
